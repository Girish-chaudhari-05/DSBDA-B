```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('forestfires.csv')
```

```python
df
```

```python
#a. Scatter Plot- Temperature vs Area Burned 
plt.figure(figsize=(8, 5))
plt.scatter(df['temp'], df['area'], alpha=0.5, color='firebrick')
plt.title('Temperature vs Area Burned')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Area Burned (ha)')
plt.grid(True)
plt.show()
```

```python
#b. Bar Plot- Number of Fires per Month 
month_counts = df['month'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
month_counts.plot(kind='bar', color='forestgreen')
plt.title('Number of Fires per Month')
plt.xlabel('Month')
plt.ylabel('Number of Fires')
plt.xticks(rotation=45)
plt.show()
```

```python
#c. Histogram- Distribution of Wind Speed 
plt.figure(figsize=(8, 5))
plt.hist(df['wind'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Wind Speed')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#d. Pie Chart- Fires by Day Type 
day_counts = df['day'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(day_counts, labels=day_counts.index, autopct='%1.1f%%', colors=['gold', 'lightcoral'])
plt.title('Fires by Day Type')
plt.show()
```

```python
#e. Line Chart- Average Temperature per Month
avg_temp_month = df.groupby('month')['temp'].mean().sort_index()

plt.figure(figsize=(8, 5))
avg_temp_month.plot(kind='line', marker='o', color='dodgerblue')
plt.title('Average Temperature per Month')
plt.xlabel('Month')
plt.ylabel('Average Temperature (Â°C)')
plt.grid(True)
plt.show()
```
