```python
# Select pollutant columns that actually exist in your dataset
pollutant_cols = ['so2', 'no2', 'rspm', 'spm', 'pm2_5']
pollutants = df[pollutant_cols].copy()
pollutants
```

```python

# Drop rows with all NaN in pollutant columns
pollutants = pollutants.dropna(how='all')
pollutants
```

```python
# Optionally fill missing values (e.g., with column means)
pollutants = pollutants.fillna(pollutants.mean())
pollutants
```

```python
# 1. Min-Max Normalization
min_max_scaler = MinMaxScaler()
min_max_normalized = pd.DataFrame(min_max_scaler.fit_transform(pollutants), columns=pollutant_cols)

```

```python
# 2. Z-Score Normalization
z_score_scaler = StandardScaler()
z_score_normalized = pd.DataFrame(z_score_scaler.fit_transform(pollutants), columns=pollutant_cols)
```

```python
# 3. Decimal Scaling Normalization
def decimal_scaling(df):
    scaled_df = df.copy()
    for col in df.columns:
        max_abs = df[col].abs().max()
        j = len(str(int(max_abs))) if max_abs != 0 else 1
        scaled_df[col] = df[col] / (10 ** j)
    return scaled_df

decimal_scaled = decimal_scaling(pollutants)

```

```python
print("\nMin-Max Normalized:\n", min_max_normalized.head())
print("\nZ-Score Normalized:\n", z_score_normalized.head())
print("\nDecimal Scaled:\n", decimal_scaled.head())
```
