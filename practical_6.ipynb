```python
# Apply Seaborn theme
sns.set_theme(style="whitegrid")
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='age', y='thalachh', hue='sex', palette='coolwarm')
plt.title('Age vs Max Heart Rate (thalach)')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.legend(title='Heart Disease')
plt.show()
```

```python
# b. ðŸ“¦ Box Plot: Cholesterol by Heart Disease
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='age', y='chol', palette='Set2')
plt.title('Cholesterol by Heart Disease Status')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Cholesterol')
plt.show()
```

```python
# c. ðŸ“Š Histogram: Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=15, kde=True, color='steelblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# d. ðŸ¥§ Pie Chart: Proportion of Heart Disease Cases
plt.figure(figsize=(6, 6))
labels = ['No Disease', 'Heart Disease']
sizes = df['output'].value_counts().sort_index()
colors = ['lightgreen', 'salmon']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Heart Disease Proportion')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```
