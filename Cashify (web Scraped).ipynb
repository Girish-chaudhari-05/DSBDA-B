```python
import requests
import bs4
```

```python
response = requests.get("https://www.cashify.in/buy-refurbished-mobile-phones/renewed-apple-iphone-13-pro/sale");
response
```

```python
soup = bs4.BeautifulSoup(response.text)
soup
```

```python
product_name = soup.find('h3',{'class':'h3 line-clamp-2'}).get_text();
print(product_name)
```

```python
product_price = soup.find("span", {"class" : "h1"}).get_text();
print(product_price)
```

```python
product_discount = soup.find("div", {"class" : "h1 text-error"}).get_text();
print(product_discount)
```

```python
reviews = soup.find_all('div', {'class': "flex flex-col py-4 sm:py-7"})
# Print reviews with their indexes
for index, review in enumerate(reviews):
    print(f"Review {index}: {review.get_text()}+'\n\n")
```

```python
avg_rating = soup.find("div", {"class" : "inherit extraFont1 bg-success p-3 py-2 sm:p-2 text-success-text-contrast rounded-lg"}).get_text()
print(avg_rating)
```

```python
total_reviews = soup.find("span", {"class" : "caption mt-1 text-surface-text"}).get_text()
print(total_reviews)
```

```python
customer_names = soup.find_all("h6", {"class": "subtitle1 text-surface-text"})
for customer_name in customer_names:
    print(customer_name.get_text() + '\n') 
```

```python
product_details = soup.find_all("div", {"class": "flex flex-col w-full"})
for product_detail in product_details:
    print(product_detail.get_text() + '\n\n') 
```

```python
top_specs = soup.find_all("div", {"class": "flex flex-wrap flex-row"})
for top_spec in top_specs:
    print(top_spec.get_text() + '\n\n') 
```

```python
image_details = soup.find_all("img", {"class": "w-20 h-20 flex-shrink-0 cursor-pointer bg-primary-bg p-2.5 object-contain border-2 rounded-lg border-surface-border"})  
for image_detail in image_details:
    print(image_detail.get('src') + '\n')
```
