```python
import pandas as pd
import numpy as np
# from sklearn.datasets import load_heart_disease
```

```python
df = pd.read_csv("C:/Users/Shubham/Desktop/dsbda/heart.csv")
```

```python
# 1. Min-Max Normalization
min_max_df = (df - df.min()) / (df.max() - df.min())
```

```python
min_max_df
```

```python
# 2. Z-Score Normalization
z_score_df = (df - df.mean()) / df.std()
```

```python
z_score_df
```

```python
# 3. Decimal Scaling Normalization
decimal_scaling_df = df.copy()
for col in df.columns:
    max_val = df[col].abs().max()
    j = len(str(int(max_val)))
    decimal_scaling_df[col] = df[col] / (10 ** j)
```
